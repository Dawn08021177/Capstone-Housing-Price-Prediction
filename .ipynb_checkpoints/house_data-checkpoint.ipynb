```python
#Load dataset
import pandas as pd
df = pd.read_csv('/mnt/c/Users/hitga/Capstone/HouseTS Dataset/Capstone-Housing-Price-Prediction/HouseTS.csv')
```

```python
#Display first few rows
df.head()
```

```python
#Check for Missing Values
df.isnull().sum()
```

```python
#Get Summary Statistics
df.describe()
```

```python
#Check Data Types
df.dtypes
```

```python
#Convert date from object to datetime
df['date'] = pd.to_datetime(df['date'])

#Convert Zipcode
df['zipcode'] = df['zipcode'].astype(str)
```

```python
#Remove for outliers
from scipy import stats

z_scores = stats.zscore(df.select_dtypes(include=['float64', 'int64']))
df = df[(abs(z_scores) < 3).all(axis=1)]
```

```python
#Feature Engineering - Create price per square foot column and month/year columns from date

df['price_per_sqft'] = df['median_sale_price'] / df['median_ppsf']

#Extract year and month
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
```

```python
#Check for duplicates
duplicates = df.duplicated().sum
print(f"Number of duplicate rows: {duplicates}")
```

```python
#Verify data consistency - checking for zero or negative values
print((df[['median_sale_price', 'median_ppsf', 'price']] <= 0).sum())
```

```python
#Drop missing values from median_ppsf column
df = df.dropna(subset=['median_ppsf'])
print(df['median_ppsf'].isnull().sum())  # Should print 0
```

```python
#check Categorical data

print(df['city'].unique())
print(df['zipcode'].nunique())
print(df['bank'].value_counts())
```

```python
#Map city code to city names
# Create a dictionary mapping city codes to full names
city_map = {
    'ATL': 'Atlanta',
    'ATX': 'Austin',
    'BOS': 'Boston',
    'BWI': 'Baltimore-Washington',
    'CHI': 'Chicago',
    'CIN': 'Cincinnati',
    'CLT': 'Charlotte',
    'DAL': 'Dallas',
    'DC': 'Washington DC',
    'DEN': 'Denver',
    'DET': 'Detroit',
    'HOU': 'Houston',
    'LA': 'Los Angeles',
    'LV': 'Las Vegas',
    'MIA': 'Miami',
    'MSP': 'Minneapolis-St. Paul',
    'NY': 'New York',
    'ORL': 'Orlando',
    'PDX': 'Portland',
    'PGH': 'Pittsburgh',
    'PHL': 'Philadelphia',
    'PHX': 'Phoenix',
    'RIV': 'Riverside',
    'SA': 'San Antonio',
    'SAC': 'Sacramento',
    'SD': 'San Diego',
    'SEA': 'Seattle',
    'SF': 'San Francisco',
    'STL': 'St. Louis',
    'TPA': 'Tampa'
}

# Map city codes to full names in a new column
df['city_full_name'] = df['city'].map(city_map)

# Verify the mapping
df[['city', 'city_full_name']].head()
```

```python
#Overwrite the original city column with full city names
df['city'] = df['city'].map(city_map)

# Check the update
df[['city']].head()
```

```python
#Verify if any city code is unmapped
df[df['city'].isnull()]['city']

```

```python
#Label encode zip code, One hot encode city
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

# Label encode zipcode
le = LabelEncoder()
df['zipcode_encoded'] = le.fit_transform(df['zipcode'])

# One-hot encode city
city_ohe = pd.get_dummies(df['city'], prefix='city')

# Combine features
features = pd.concat([df.drop(columns=['zipcode', 'city']), city_ohe], axis=1)



```

```python
# Check the first few rows to see the new encoded columns
print(df[['zipcode', 'zipcode_encoded', 'city']].head())

# Check the one-hot encoded city columns shape and sample columns
print(city_ohe.head())

# Check the combined features shape to confirm dimensions
print(features.shape)

# Check column names of features (optional)
print(features.columns.tolist())
```
